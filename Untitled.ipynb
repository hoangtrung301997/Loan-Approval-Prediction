```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import seaborn as sns
```

```python
data = pd.read_csv("credit_risk_dataset.csv")
```

```python
data.describe()
```

Numerize categorical data

```python
data.head()
```

```python
#encode person_home_ownership
encoder = OneHotEncoder(sparse=False)
encoded_home_owner = encoder.fit_transform(data[['person_home_ownership']])
encoded_data_home_owner = pd.DataFrame(encoded_home_owner, columns = encoder.get_feature_names_out(['person_home_ownership']))

#encoded loan_intent
encoded_loan_intent = encoder.fit_transform(data[['loan_intent']])
encoded_data_loan_intent = pd.DataFrame(encoded_loan_intent, columns = encoder.get_feature_names_out(['loan_intent']))

data = pd.concat([data, encoded_data_home_owner, encoded_data_loan_intent], axis=1)
```

```python
#encode loan_grade
data['loan_grade'] = pd.Categorical(data.loan_grade).codes

#encode previous_default
data['cb_person_default_on_file'] = data['cb_person_default_on_file'].map({'Y':1,'N':0})
```

# Visualize key statistic

```python
data = data.drop(['person_home_ownership', 'loan_intent'], axis=1)
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.title('Box plot of Features')
plt.show()
```

Need to standardize to bring features to same scale

```python
scaler = StandardScaler()
df = scaler.fit_transform(data)
df = pd.DataFrame(data, columns = data.columns)

```

# Correlation Matrix

```python
corr_matrix = df.corr()

#Visualize correlation matrix
plt.figure(figsize=(16,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation matrix")
plt.show()
```

```python
# Define a correlation threshold (e.g., 0.7)
threshold = 0.7

# Create a mask to remove upper triangle of the correlation matrix
mask = (corr_matrix.abs() > threshold) & (corr_matrix.abs() < 1)

# Get pairs of highly correlated features
high_corr_pairs = [(col1, col2) for col1 in corr_matrix.columns for col2 in corr_matrix.columns 
                   if mask.loc[col1, col2]]

# Print highly correlated feature pairs
print("Highly Correlated Feature Pairs (above threshold):")
for pair in high_corr_pairs:
    print(pair)
```
